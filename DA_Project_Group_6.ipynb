# Hypothesis Analysis 

## Hypothesis - 1

Null hypothesis (H0): There is no significant difference in mean Big Five scores between males and females.

Alternative hypothesis (HA): There is a significant difference in mean Big Five scores between males and females.

```python
data = df_big5.drop(['Filename', 'Name'],  axis=1)
male_data = data[data['Gender'] == 'm']
female_data = data[data['Gender'] == 'f']
```

```python
from scipy.stats import zscore, norm

# Perform z-tests to compare the means of each personality trait between males and females
openness_z = (male_data['Openness'].mean() - female_data['Openness'].mean()) / (data['Openness'].std(ddof=1) / len(data)**0.5)
openness_p = 2 * (1 - norm.cdf(abs(openness_z)))
conscientiousness_z = (male_data['Conscientiousness'].mean() - female_data['Conscientiousness'].mean()) / (data['Conscientiousness'].std(ddof=1) / len(data)**0.5)
conscientiousness_p = 2 * (1 - norm.cdf(abs(conscientiousness_z)))
extraversion_z = (male_data['Extraversion'].mean() - female_data['Extraversion'].mean()) / (data['Extraversion'].std(ddof=1) / len(data)**0.5)
extraversion_p = 2 * (1 - norm.cdf(abs(extraversion_z)))
agreeableness_z = (male_data['Agreeableness'].mean() - female_data['Agreeableness'].mean()) / (data['Agreeableness'].std(ddof=1) / len(data)**0.5)
agreeableness_p = 2 * (1 - norm.cdf(abs(agreeableness_z)))
neuroticism_z = (male_data['Neuroticism'].mean() - female_data['Neuroticism'].mean()) / (data['Neuroticism'].std(ddof=1) / len(data)**0.5)
neuroticism_p = 2 * (1 - norm.cdf(abs(neuroticism_z)))

# Print the results
print(f"Openness z-value: {openness_z:.3f}, p-value: {openness_p:.3f}")
print(f"Conscientiousness z-value: {conscientiousness_z:.3f}, p-value: {conscientiousness_p:.3f}")
print(f"Extraversion z-value: {extraversion_z:.3f}, p-value: {extraversion_p:.3f}")
print(f"Agreeableness z-value: {agreeableness_z:.3f}, p-value: {agreeableness_p:.3f}")
print(f"Neuroticism z-value: {neuroticism_z:.3f}, p-value: {neuroticism_p:.3f}")
```

```python
# let the value of significance label ia 0.05
alpha = 0.05
# cases = ['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']
for p_value in [openness_p, conscientiousness_p, extraversion_p, agreeableness_p,  neuroticism_p]:
      if p_value <= alpha:
          print(f"Reject the null hypothesis. H(A) is selected.")
      else:
          print(f"Accept the null hypothesis. H(0) is selected.")
```

## Hypothesis - 2

Null hypothesis (H0): There is a significant relation between words related to anxiety with security.

Alternative hypothesis (HA): There is no significant relation between words related to anxiety qith security.

```python
data_hypo_2=df;
anxiety_data = data_hypo_2['anx']
security_data = data_hypo_2['Security']
```

```python
#from scipy.stats import zscore, norm
import scipy.stats as stats

r,p_value = stats.pearsonr(anxiety_data, security_data)

print('Correlation coefficient (r):', r)
print('p-value:', p_value)
```

```python
# let the value of significance label ia 0.05
alpha = 0.05
if p_value <= alpha:
        print(f"Reject the null hypothesis. H(A) is selected.")
else:
        print(f"Accept the null hypothesis. H(0) is selected.")
```
